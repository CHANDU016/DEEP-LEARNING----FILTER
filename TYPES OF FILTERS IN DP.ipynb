## TYPES OF FILTERS

 Smoothing (Blurring) Filters

```python
import cv2
```

```python
image = cv2.imread('ssss.jpeg')
```

```python
plt.imshow(image)
```

```python
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_rgb)
```

```python
blurred_image= cv2.GaussianBlur(image_rgb,(5,5),0)
```

```python
import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(blurred_image)
plt.title('Gaussian Blurred Image')

plt.show()
```

##  Edge Detection Filters

```python
# Apply Sobel filter (for edge detection)
sobel_x = cv2.Sobel(image_rgb, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobel_y = cv2.Sobel(image_rgb, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges
```

```python
# Convert back to 8-bit for display
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
```

```python
# Combine both Sobel filters
sobel_combined = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
```

```python
# Display the edge-detected images
plt.subplot(1, 3, 1)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')

plt.subplot(1, 3, 2)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')

plt.subplot(1, 3, 3)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Combined')
```

## Sharpening Filter

```python
# Create a sharpening kernel

import numpy as np
kernel = np.array([[0, -1, 0], 
                   [-1, 5, -1], 
                   [0, -1, 0]])
```

```python
# Apply the sharpening filter
sharpened_image = cv2.filter2D(image_rgb, -1, kernel)
```

```python
# Display the sharpened image
plt.imshow(sharpened_image)
plt.title('Sharpened Image')
plt.show()



```
